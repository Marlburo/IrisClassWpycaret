**Iris classification using sklearn, pandas, numpy and pycarett**

```python
!pip install pycaret shub
import numpy as np
import pandas as pd
import matplotlib as plt
from pycaret.classification import *
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
t_names = ['setosa', 'versicolor', 'virginica']
iris_class = setup(iris_df,target='target')
```

```python
models()
```

```python
best_model = compare_models()
```

```python
best_model
```

```python
predict_model(best_model,iris_df.tail())
```

```python
save_model(best_model, model_name='quadratic-discriminant-analysis')
```
